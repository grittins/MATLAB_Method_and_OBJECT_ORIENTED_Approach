```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create a plot with ax.plt()
fig, ax = plt.subplots()

# Create the plot
ax.plot(x_axis, y_axis, marker="D", label="Boston", color="green", linewidth=2)

# Add a title and axes labels
ax.set_title("PyBer Fare by Month")

# Add a y-axis limit
ax.set_ylim(0,45)

# Add grid lines
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o') # adding 'o' creates a scatter plot instead of a line chart

# OR just use the following for same results
# plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, c="red", label="Chicago")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
# to increase the size of the data points
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#Add a black edge color to the circles.
#Make the linewidth of the circles 2 points.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           c= "xkcd:sky blue", 
           label="Boston", 
           linewidth = 2,  
           s=[i * 5 for i in y_axis], 
           alpha = 0.8, 
           edgecolor='0')

# Invert the y-axis so that January is at the top.
ax.invert_yaxis()

#Add a legend for the city of Boston.
#Add a title.
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)
```

```python
# Pie Chart using MATLAB method
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Pie chart using the Object Oriented Approach
fig, ax = plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%', 
    shadow=True, 
    startangle= 90, 
    counterclock = False)

plt.show()
```
