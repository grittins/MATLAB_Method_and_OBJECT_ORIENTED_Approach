```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create a plot with ax.plt()
fig, ax = plt.subplots()

# Create the plot
ax.plot(x_axis, y_axis, marker="D", label="Boston", color="green", linewidth=2)

# Add a title and axes labels
ax.set_title("PyBer Fare by Month")

# Add a y-axis limit
ax.set_ylim(0,45)

# Add grid lines
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot horizontally
plt.barh(x_axis, y_axis)
```

```python
# Change the bar colors to magenta.
# Invert the y-axis so that January is at the top.
plt.barh(x_axis, y_axis, color='magenta', label="Boston")
plt.gca().invert_yaxis()

# Add a legend for the city of Boston.
plt.legend()

# Add a title and axes labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# Create the plot for ax.bar()
fig, ax=plt.subplots()

# Change the bar colors to magenta.
# Invert the y-axis so that January is at the top.
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.invert_yaxis()

# Add a legend for the city of Chicago.
ax.legend()

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```
